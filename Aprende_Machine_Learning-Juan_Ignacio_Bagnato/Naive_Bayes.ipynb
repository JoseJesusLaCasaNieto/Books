# Naive Bayes

## 1. Los datos de entrada

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
import os
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest
```

```python
rute_data = os.path.join('..', '..', 'comprar_alquilar.csv')
dataframe = pd.read_csv(rute_data)
dataframe.head(10)
```

## 2. Visualizaci√≥n de datos

```python
print(dataframe.groupby('comprar').size())
```

```python
dataframe.drop(['comprar'], axis=1).hist()
plt.show()
```

## 3. Preparar los datos de entrada

```python
dataframe['gastos'] = (dataframe['gastos_comunes'] + dataframe['gastos_otros'] + dataframe['pago_coche'])
dataframe['financiar'] = dataframe['vivienda'] - dataframe['ahorros']
dataframe.drop(['gastos_comunes', 'gastos_otros', 'pago_coche'], axis=1).head(10)
```

```python
reduced = dataframe.drop(['gastos_comunes', 'gastos_otros', 'pago_coche'], axis=1)
reduced.describe()
```

## 4. Feature Selection o Selecci√≥n de Caracter√≠sticas

```python
X = dataframe.drop(['comprar'], axis=1)
y = dataframe['comprar']

best = SelectKBest(k=5)
X_new = best.fit_transform(X, y)
print(X_new.shape)
selected = best.get_support(indices=True)
print(X.columns[selected])
```

```python
used_features = X.columns[selected]

colormap = plt.cm.viridis
plt.figure(figsize=(12, 12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(
    dataframe[used_features].astype(float).corr(),
    linewidths=0.1,
    vmax=1.0,
    square=True,
    cmap=colormap,
    linecolor='white',
    annot=True
)
```

## 5. Crear el modelo Gaussian Naive Bayes con SKLearn

```python
# Split dataset in training and test datasets
X_train, X_test = train_test_split(dataframe, test_size=0.2, random_state=6)
y_train = X_train['comprar']
y_test = X_test['comprar']
```

```python
# Instantiate the classifier
gnb = GaussianNB()

# Train classifier
gnb.fit(
    X_train[used_features].values,
    y_train
)
y_pred = gnb.predict(X_test[used_features])

print('Precisi√≥n en el set de entrenamiento: {:.2f}'.format(gnb.score(X_train[used_features], y_train)))
print('Precisi√≥n en el set de test: {:.2f}'.format(gnb.score(X_test[used_features], y_test)))
```

## 6. Probemos el modelo: ¬øComprar o alquilar?

```python
print(gnb.predict([[2000, 5000, 0, 5, 200000],
                   [6000, 34000, 2, 5, 320000]]))
```
